```python
#Load in libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Bar chart using Hotdog Contest Winners dataset

```python
#Read in file and view example
hdwin = pd.read_excel("hotdog-contest-winners.xlsm")
hdwin.head()
```

```python
#Need to update two column names to make it easier to use
hdwin.rename(columns = {'Dogs eaten':'Dogs_eaten'}, inplace = True)
hdwin.rename(columns = {'New record':'New_record'}, inplace = True)
hdwin.head()
```

```python
#Bar chart
plt.bar(hdwin.Year, hdwin.Dogs_eaten, color = 'blue')
plt.xlabel("Year")
plt.ylabel("Number of dogs eaten by winner")
plt.title("Number of hotdogs eaten by the contest winner")
plt.show()
```

## Stacked bar chart

```python
# For this exercise I have modified the hotdog-places file
hdp = pd.read_excel("hotdog-places.xlsm")
hdp.head()
```

```python
#Stacked bar charts
plt.bar(hdp.Year, hdp.Chicago, label = '#Chicago', color = 'green')
plt.bar(hdp.Year, hdp['New York'], label = '#New York', color = 'red')
plt.bar(hdp.Year, hdp.California, label = '#California', color = 'yellow')
plt.xlabel("Year")
plt.ylabel("Count of hotdog restorents")
plt.legend(loc = 'upper left')
plt.xticks(rotation = 90, fontsize = 10)
plt.title('Hotdog places groath chart by city')
plt.show()
```

## Pie chart using Obama Approval Ratings dataset

```python
#Read in dataset and get preview
hcw = pd.read_excel("hotdog-contest-winners.xlsm")
hcw.head()
```

```python
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = hcw.Country
sizes = hcw['Dogs eaten']

# explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
# , explode=explode
fig1, ax1 = plt.subplots(figsize=(20, 25))
ax1.pie(sizes, labels=labels, autopct=autopct_format(sizes),
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('hotdogs eaten Winner by country')
plt.show()
```

## Donut chart using Hot dogs eaten winner

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = hcw.Country
sizes = hcw['Dogs eaten']

fig, ax1 = plt.subplots(figsize=(20, 22))
ax1.pie(sizes, labels=labels, autopct=autopct_format(sizes),
        shadow=True, startangle=90)

#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.65,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('hotdogs eaten Winner by country')
plt.show()
```
